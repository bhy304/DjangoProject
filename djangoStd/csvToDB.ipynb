```python
%%writefile DBMgr.py
import pymysql
import pandas as pd

class DBHelper:
    conn = None    
    def __init__(self):
        self.db_connect()
        #DBHelper 생성될 때 DB 자동 연결
    
    #DB 연결
    def db_connect(self):
        self.conn = pymysql.connect(host="localhost", user="root", password="1234",db="student", charset="utf8")
    
    #DB 연결 해제
    def db_disconnect(self):
        if self.conn:
            self.conn.close()
            
    def selectMajor(self):
        with self.conn.cursor() as cur:
            sql = "select * from student_major"
            cur.execute(sql) #sql 실행
            rows = cur.fetchall()
        return rows
        
    def insert(self):
        with self.conn.cursor() as cur:
            csv_data = pd.read_csv('csv/major.csv')
            sql = "INSERT INTO student_major(major_id, major_title) values(%s, %s)"
            for row in csv_data.get_values():
                cur.execute(sql, tuple(row))
        self.conn.commit()
        
#현재 class가 단독으로 실행할 때만 수행한다.(테스트용)
if __name__ == '__main__': 
    db = DBHelper()
    db.insert()
    print(db.selectAll())
    db.db_disconnect()
```
